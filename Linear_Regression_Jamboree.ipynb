### 4.	Testing the assumptions of linear regression model

#### a. Multicollinearity check by VIF score

```python
def vif(df1):
    vif_d = pd.DataFrame()
    vif_d['feature'] = df1.columns

    vif_d['VIF'] = [variance_inflation_factor(df1.values, i) for i in range(len(df1.columns))]

    return vif_d
```

```python
res = vif(df.iloc[:, :-1])
res
```
